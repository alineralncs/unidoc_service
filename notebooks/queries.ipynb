```python
# precision

def precision_at_k(k, expected_results, lsa_recommendations):
    precision = 0
    for i, recommendations in enumerate(lsa_recommendations):
        expected = expected_results[str(i+1)]
        expected = set(expected)
        recommendations = set(recommendations[:k])
        precision += len(expected.intersection(recommendations)) / k
    return precision / len(lsa_recommendations)

for k in [5, 10]:
        precision = precision_at_k(k, expected_results, df_lsa_recommendations['lsa_recommendations'])
        print(f"Precision at {k}: {precision}")

# recall 

def recall_at_k(k, expected_results, lsa_recommendations):
    recall = 0
    for i, recommendations in enumerate(lsa_recommendations):
        expected = expected_results[str(i+1)]
        expected = set(expected)
        recommendations = set(recommendations[:k])
        recall += len(expected.intersection(recommendations)) / len(expected)
    return recall / len(lsa_recommendations)

print(recall_at_k(5, expected_results, df_lsa_recommendations['lsa_recommendations']))
print(recall_at_k(10, expected_results, df_lsa_recommendations['lsa_recommendations']))
```

```python
def ndcg_at_k(k, expected_results, lsa_recommendations):
    ndcgs = []
    for i, recommendations in enumerate(lsa_recommendations):
        expected = expected_results[str(i+1)]
        expected = set(expected)
        recommendations = recommendations[:k]
        dcg = 0
        for j, item in enumerate(recommendations):
            if item in expected:
                dcg += 1 / (j + 1)
        idcg = sum([1 / (j + 1) for j in range(min(len(expected), k))])
        ndcgs.append(dcg / idcg)
    return sum(ndcgs) / len(ndcgs)

for k in [5, 10]:
    ndcg = ndcg_at_k(k, expected_results, df_lsa_recommendations['lsa_recommendations'])
    print(f"NDCG at {k}: {ndcg}")

def ndcg_at_k_each_result(k, expected_results, lsa_recommendations):
    ndcgs = []
    for i, recommendations in enumerate(lsa_recommendations):
        expected = expected_results[str(i+1)]
        expected = set(expected)
        recommendations = recommendations[:k]
        dcg = 0
        for j, item in enumerate(recommendations):
            if item in expected:
                dcg += 1 / (j + 1)
        idcg = sum([1 / (j + 1) for j in range(min(len(expected), k))])
        ndcgs.append(dcg / idcg)
    return ndcgs

ndcg_5 = ndcg_at_k_each_result(5, expected_results, df_lsa_recommendations['lsa_recommendations'])
ndcg_10 = ndcg_at_k_each_result(10, expected_results, df_lsa_recommendations['lsa_recommendations'])

plt.hist(ndcg_5, bins=20, alpha=0.5, label='NDCG@5')
plt.hist(ndcg_10, bins=20, alpha=0.5, label='NDCG@10')
plt.legend(loc='upper right')
plt.show()
```
