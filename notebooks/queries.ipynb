```python
import requests

def get_data_query_rs(query: str):
    url = f'http://127.0.0.1:8049/documents/getDocumentQuery/{query}'

    response = requests.get(url)

    data = response.json()
    return data
```

```python
from prettytable import PrettyTable
def print_pretty_table(df):
    table = PrettyTable()
    table.field_names = ["Query", "LSA Recommendations"]

    for index, row in df.iterrows():
        recommendations = "\n ".join(row['lsa_recommendations'] ) 

        table.add_row([row['Query'], recommendations])
        table.add_row(["", "" ])


    print(table)
```

#### Consulta com Relações Semântica

```python
import pandas as pd
import pprint

query_results = [get_data_query_rs(queries_dict[str(i)]) for i in range(1, 38)]

# Criação do dataframe
data = {
    'Query': list(queries_dict.values()),
    'lsa_recommendations': [result['lsa_recommendations'] for result in query_results]
}

df_lsa_recommendations = pd.DataFrame(data)



```

```python
print_pretty_table(df_lsa_recommendations)
```
